<h1><b> O Ciclo Rankine </b></h1>

<p style="text-align: justify"> O Ciclo Rankine opera a partir da geração do energia pela movimentação de turbinas com pressão de vapor, essa movimentação gera energia cinética que posteriormente é convertida em energia elétrica.</p>

<p style="text-align: justify"> O funcionamento do ciclo depende da queima de combustíveis para geração de energia, desta forma, a energia fornecida pela combustão vaporiza a água, engatilhando o ciclo. A base principal de um ciclo termodinamico é que a água pode ser vaporizada e condensada diversas vezes, para propagar a energia fornecida pelo sistema. O rendimento do ciclo é medido pela diferença entre a energia fornecida e a gerada, levando em conta também possíveis gastos durante o ciclo. Desta forma, o objetivo deste estudo, é análisar cada etapa do Ciclo Rankine afim de avaliar o seu rendimento, com base na combustão dos combustiveis trabalhados anteriormente (Desafios 1, 2 e 3). </p>

<p style="text-align: justify"> É importante ressaltar que a água não é o único fluido de trabalho do Ciclo (seja ele inorgânico ou orgânio), mas acaba sendo muito utilizada devido suas propriedades químicas favoráveis, como por exemplo: não ser tóxica, reativa, ser abundante e ter baixo custo, além claro, de ter o calor específico alto ao ser comparado com outras substâncias comuns, sendo este de 4,19 kJ/kg K. Outro tópico importante é que no ciclo de Rankine, a Lei dos Gases Ideais quase não pode ser usado (o vapor não segue pV=nRT). Portanto, todos os parâmetros importantes de água e vapor são tabulados nas chamadas “ Tabelas de vapor ”, neste experimento, iremos utilizar as informações do database NIST.

<h2><b> Etapas do Ciclo Rankine </b></h2> 

<p style="text-align: justify">O ciclo de Rankine ideal  é dividido em quatro etapas - dois processos isentrópicos (adiabáticos) alternados com dois processos isobáricos, que envolvem alterações de temperatura, pressão e fase. As operações se iniciam na Bomba, visto que ela fornece a energia para a água estar em estado líquida na Caldeira, onde se é fornecida energia térmica para aquecer a água líquida até o ponto de vapor supersaturado (temperatura acima do ponto de ebulição). Uma vez que a água é completamente ebulida, a pressão do sistema aumenta, bem como a entropia, levando a terceira etapa do sistema, onde o vapor em alta pressão movimenta a Turbina, gerando energia em forma de trabalho. Após passar pela Turbina, o vapor é encaminhado para um Condensador, com temperatura e pressão inferiores à etapa anterior, estando agora em fase líquido + vapor. Essa transferência, contudo, acontece de forma isoentrópica, portanto, a entropia da água ao entrar no Condensador é igual a entropia da água ao sair da Caldeira. No Condensador, a entropia será diminuída, afim de liquefazer totalmente a água, antes de encaminhá-la para a próxima etapa, a Bomba. Na Bomba, é fornecida energia, para que ela impulsione a água em estado líquido novamente para a Caldeira, onde se reinicia o ciclo. </p>
<p style="text-align: justify"> Ademais, ressalta-se que uma das maiores vantagens do Ciclo de Rankine é que o processo de compressão da bomba ocorre com a água em um estado líquido e como citado, para isso ocorrer, deve-se condensar o vapor do fluído para o líquido e isso tem como consequência, uma pressão reduzida na turbina, impactando em uma baixa energia necessária para a bomba de alimentação, consumindo apenas de  1% a 3% da potência de saída da turbina - todos esses processos contribuem para uma alta eficiência do ciclo. </p>

<p style="text-align: justify"> Desta forma, a geração de energia pelo Ciclo de Rankine se dá pelo trabalho gerado na movimentação da Turbina. Para avaliar a eficiência do ciclo, contudo, é preciso considerar as energias fornecidas para o sistema, para que sejam descontadas nas variações de energia do sistema em todas as etapas. Sabemos que é necessário fornecer uma quantidade inicial de energia, que será absorvida em forma de calor na Caldeira, esta energia é definida como $Q_{h}$, já na etapa 2, temos energia sendo gerada em forma de trabalho, $W_{turbina}$. No condensador há uma perda de energia térmica dada pelo resfriamento da água $Q_{c}$, e, por fim, temos o trabalho realizado pela Bomba $W_{bomba}$. Sendo assim, a taxa de eficiência do sistema $e$ é definida como uma razão entre o trabalho líquido, aquele que pode ser de fato beneficiado ao fim do ciclo, e a energia gasta:</p>

<p style="text-align: center">  $$ e = \frac{W_{líquido}}{Q_{h}} = \frac{W_{turbina}- W_{bomba}}{Q_{h}}$$ </p> 

<p style="text-align: justify"> Podemos analisar, portanto, que o ponto crucial do Ciclo de Rankine, é o trabalho gerado na Turbina, decorrente da pressão exercida pela vapor. Esta, por sua vez, depende diretamento da quantidade de calor fornecido para a Caldeira, e sua capacidade de ebulir a água. Por isso, ao selecionar a mistura que será utilizada como combustível em uma termoelétrica, deseja-se compostos com alta entalpia de combustão, pois é a partir da queima destes combustíveis que será fornecida a energia necessaria para que o sistema opere de forma eficiente.

<h3> <b> Vamos definir os parâmetros do nosso Ciclo? Parte I </b> </h3> 

<p style="text-align: justify"> Uma medida muito importante para a determinação do funcionamento do nosso ciclo é categorizar suas temperaturais iniciais e finais, além de definir o fluxo de massa de água que irá percorrer o sistema. Essas parametrizações nos permitem definir a capacidade e eficiência do Ciclo de Rankine. Como podemos definir esses parâmetros? </p>
<p style="text-align: justify"> Como discutido anteriormente, sabemos que os ciclos de potência a vapor, como o Ciclo de Rankine Ideal, operam de modo geral com o bombeamento de água a alta pressão de modo que esse fluído vá para uma caldeira, onde o calor é adicionado a pressão constante, em seguida,  esse fluído se torna um vapor que atinge uma turbina, onde expande adiabaticamente, esfriando o fluído e deixando-o na baixa pressão original. Por fim, o fluido é totalmente condensado. </p>
<p style="text-align: justify"> Por meio desse processo, conseguimos observar que a água sofre com grandes diferenças de temperaturas e existe duas fórmulas muito utilizadas no estudo da termodinâmica que nos permite identificar informações como massa e temperatura, sendo esta: a equação do calor sensível e a equação do calor latente.  </p>
<h3> <b> Calorimetria </b> </h3> 
<p style="text-align: justify"> O calor específico de uma substância define-se como a quantidade de calor que entra ou que sai de uma unidade de massa, ao variar em 1 (um) grau a temperatura da substância. Sua postulação é dada por:

$$ Cp = ( \frac{dQ}{ m \cdot dT } )  $$
<p style="text-align: justify">  Isolando essa equação e considerando uma unidade de massa qualquer, temos a equação do calor sensível que nos relata aquantidade de calor transferida a um sistema durante um processo em que ocorre uma variação de temperatura, sem que haja mudança de fase. Logo, temos:</p>
 </p>
$$ dQ = (Cp \cdot  m ) dt $$
$$ Q = m \cdot \int_{Ti}^{Tf} Cp(T) \, dT $$
<p style="text-align: justify"> sendo que: </p>

- Q é o calor sensível transferido para o sistema (em kg/Joules);
- ∫ representa a integral definida;
- m é a massa do sistema (em kg);
- Cp é a capacidade térmica do sistema em pressão constante(J/Kg);
- dT é a variação de temperatura (em graus Kelvin) ao longo do processo de transferência de calor. </p>

De modo a se obter a massa, temos que isolar a componente de massa da equação. Assim:
$$ m = \frac{Q}{\int C(T) dT}$$

<p style="text-align: justify"> Essa forma nos permite calcular a massa do sistema quando conhecemos o calor sensível transferido (Q) e  quando se é possívrl realizar a integração da capacidade térmica (C) em relação à variação de temperatura (dT). Entretanto, a fórmula do calor sensível não pode ser a única a ser considerada para a determinação do projeto, visto que há de se considerar o calor latente do sistema quando há mudança de fase da água, como por exemplo, quando ela está no estado líquido e vai para o estado de vaporização.  </p>

<p style="text-align: justify">  O cálculo do calor latente envolve a multiplicação da massa pelo calor latente específico do material, sendo este, a quantidade de calor necessária para que uma substância sofra uma mudança de fase, sua fórmula é dada por: </p>

$$ Q = m \cdot L $$

<p style="text-align: justify"> sendo que: </p>

- Q é o calor latente (em Joules),
- m é a massa do material (em kg),
- L é o calor latente específico (em J/kg).

<p style="text-align: justify">  Especificamente, no caso do Calor Latente, não iremos obter uma lista com as variações da constante específica do calor latente, já que em sua fórmula não há a variação de temperatura como o do calor sensível, logo, não há a necessidade de realizar o método de integração para a obtenção dessa informação. Iremos utilizar o dado tabelado de 2256 J/kg, desse modo, temos como fórmula utilizada:  </p>
$$ Q = m \cdot 2256 $$

<p style="text-align: justify">  Em um primeiro momento, iremos realizar o plot das informações dos estados da água pela temperatura (líquida e vapor), determinando a área dessas para enfim, realizar o cálculo da massa de água que é obtida pela divisão do calor Q pela integral dos estados de água do calor sensível mais o calor latente.
```python
# Função da massa
def massa_agua(Q, calor_liquido, calor_vapor, calor_latente, T):
    ''' Função que calcula a massa de água que irá ser utilizada no Ciclo de Rankine. Iremos realizar a soma do calor sensível e do calor latente do sistema, a fim de se obter a massa.
    Args: Q(energia total do sistrema), Calor Sensível Liquido e de Vaporização, Temperaturas utilizadas, Calor Latente.
    Return: a massa de água do fluxo.
    '''
    integral_agua = integrate.trapezoid(calor_liquido, T) 
    integral_vapor = integrate.trapezoid(calor_vapor, T)
    m_agua = Q / (integral_agua + integral_vapor + calor_latente)
    print(f'A massa de água que será utilizada no Ciclo de Rankine é de {m_agua} kg.')
```

```python
# Vamos calcular as massas? Para isso, temos que definir algumas constantes
# Constantes
calor_latente = 2256 #J/kg
calorzin_Q = 3.3463e7 #J/kg

# Apresentar o resultado
massa_agua(calorzin_Q, Cp_liq, Cp_vap,calor_latente, T)
```

<p style="text-align: justify"> Obtidos todos os valores e configurações da biblioteca, o que fazemos com essas informações? Quais são os cálculos que devemos realizar para poder obter todas as informações de calor e trabalho? Pois bem, os cálculos são obtidos por meio da 1 ª Lei da Termodinâmica, isso mesmo, o famoso princípio de conservação de energia, o qual afirma que a energia total de um sistema permanece constante, isto é, a energia não pode ser criada nem destruída, apenas transformada de uma coisa para outra. Essa Lei pode ser expressa matemáticamente por: </p>

$$ dU = dQ - dW $$

<p style="text-align: justify">  sendo: </p>

- dU é a energia interna do sistema,
- dQ é o calor transferido para o sistema durante uma transformação termodinâmica,
- dW é o trabalho realizado pelo sistema durante uma transformação termodinâmica.


<h3> <b> Aplicando a 1ª Lei da Termodinâmica </b> </h3> 
<h3> <b> Bomba </b> </h3> 
<p style="text-align: justify"> Relembrada a 1ª Lei da Termodinâmica, vamos analisa-la agora em termos de Entalpia com o intuito de simplificar as postulações matemáticas do trabalho realizado pela bomba. Essa análise pode ser feita em sistema em que há variação de volume do sistema a pressão constante, como o caso do Ciclo de Rankine. Para realizar essa análise, considera-se um sistema em que ocorre apenas trabalhos de expansão ou compreensão, assim o trabalho realizado é dado por: </p>
    
$$ dW = -PdV$$
<p style="text-align: justify">  sendo: </p>

- P a pressão do sistema,
- dV é o trabalho realizado pelo sistema.
 
 <p style="text-align: justify"> Dada a definição de dW, substitui-se essa definição na 1ª Lei da Termodinâmica: </p>

$$ dU = dQ - (-PdV) $$
$$ dU = dQ + PdV $$

 <p style="text-align: justify"> Agora vamos expressar essa lei em termos de entalpia (H) ao invés de energia interna (U), isso é possível porque estamos considerando um sistema de pressão constante, o que nos permite definir a entalpia como a soma da energia interna do sistema e do produto da pressão pelo volume do sistema (H ≡ U + P.v) e ao considerarmos a entalpia (dH), temos então:

$$dH = dU + PdV $$

 <p style="text-align: justify"> Substituindo a expressão dU = dQ + PdV, obtemos: </p>

$$ dH = dQ + PdV $$

<p style="text-align: justify"> Obtida essa equação matemática e sabendo que na fase da bomba temos uma compreensão adiabática, logo, não há trocas de calor e de matéria entre a nossa bomba com o meio externo, o que permite com que a entropia da água seja constante, o valor de dQ = 0, portanto, a 1ª Lei da Termodinâmica é expressa por: </p>
$$ dH =  PdV $$
<p style="text-align: justify"> Sabendo que pdv é considerado o W(trabalho) ocorrido e que dH é a diferença das entalpias do processo 1-2, temos que a fórmula do trabalho da bomba será dado por: </p>
$$ W_{bomba} = h_1 - h_2 = v \cdot (P_1 - P_2) $$


```python
# Líquido Saturado, logo x = 0
p1 = 10 # kPa
h1 = mp_water.hs(p=p1)[0]
T1 = mp_water.Ts(p=p1)[0]
s1 = mp_water.ss(p=p1)[0]
p2 = 8000 # kPa
s2= s1
v = 1/mp_water.ds(p=p1)[0]
w_p = v*(p2-p1)
print(f"O trabalho requerido pela bomba é de: {round(float(w_p),1)} kJ/kg")
```

<h4> <b> Diagrama T-s do ponto da bomba </b> </h4> 


```python
# Definir os eixos do dataframe
p = np.linspace(1,22063,1000)
T = mp_water.Ts(p=p)
s = mp_water.ss(p=p)


# Ajustar o texto 
font = {'family' : 'Times New Roman',
        'size'   : 22}

# Chamar imagem 
plt.figure(figsize=(15,10))
plt.title('Diagrama T-S do Ciclo de Rankine Ideal')
plt.rc('font', **font)
plt.grid()

#Plot das informações
plt.plot(s[0],T, linestyle='solid', linewidth=3, color = 'turquoise')
plt.plot(s[1],T, linestyle='solid', linewidth=3, color = 'fuchsia')
plt.ylabel('Temperature (K)')
plt.xlabel('Entropy (kJ/kg K)')

#Marcação dos Pontos
plt.plot(s1, T1, 'o', color = 'black')

# Delimitando área
plt.xlim(0,2)
plt.ylim(318,320)

#Plotando texto
plt.text(s1-.1,T1,f' Bomba (1)\nT = {round(float(T1),2)} K\nh = {round(float(h1),1)} kJ/kg\n s = {round(float(s1),3)} kJ/kgK',
    ha='right',backgroundcolor='white')

plt.show()
```

<h3> <b> Caldeira </b> </h3> 
<p style="text-align: justify"> Depois de calculado o trabalho requerido pela bomba, pode-se calcular o valor da entalpia (ponto 2). Esse valor de entalpia irá nos permitir realizar o cálculo do calor de entrada da caldeira, ressalta-se que nessa parte do processo temos um processo isobárico, portanto o nosso trabalho será igual a zero (Vdp = 0). E se o nosso trabalho será igual a zero, utilizando a 1ª Lei da Termodinâmica, temos como formulação matemática: </p>

$$ dH = dQ + PdV $$
$$ dH = dQ  $$

<p style="text-align: justify">Sabendo que dQ diz respeito ao calor e que dH é se refere as diferenças de entalpia do Processo 2-3, temos como equação matemática que o calor da caldeira será de: </p>
$$ Q_{caldeira} = h_3 - h_2  $$

```python
h2 = h1+w_p
T2 = mp_water.T_h(p=p2,h=h2)

print(f" O valor de entalpia no Ponto 2 (h2) é de {round(float(h2),1)} kJ/kg")
```

```python
# O vapor sai da caldeira como vapor saturado, portanto x = 1.
p3 = p2
T3 = mp_water.Ts(p=p3)
h3 = mp_water.hs(p=p3)[1]
s3 = mp_water.ss(p=p3)[1]
s3dash = mp_water.ss(p=p3)[0]
T3dash = T3
q_H = h3-h2

print(f"O calor de entrada da caldeira é de {round(float(q_H),1)} kJ/kg")
```

<h3> <b> Turbina </b> </h3> 
<p style="text-align: justify"> Neste momento, a turbina apresenta um processo isentrópico (dQ = 0), não há transferência de calor. Portanto, considera-se apenas o trabalho da turbina que aplicando a 1ª Lei da Termodinâmica, resulta na equação de diferença da entalpia do ponto 3 subtraído pelo ponto 4: </p>
$$ W_{turbina} = h_3 - h_4  $$


```python
p4 = p1
s4 = s3
T4, x = mp_water.T_s(s=s4,p=p4, quality=True)
h4 = mp_water.h(p=p4,x=x)
w_t = h3-h4
print(f"A qualidade do vapor de baixa pressão é de {round(float(x),4)}")
print(f"O trabalho gerado pela turbina é de {round(float(w_t),1)} kJ/kg")
```

<h3> Eficiência de Rankine </h3>
<p style="text-align: justify"> O último processo do Ciclo é a análise de sua eficiência -  medida de quão efetivamente um sistema ou ciclo utiliza a energia disponível para realizar o trabalho de interesse.  A equação utilizada será uma já discutida no experimento, a qual é:</p>
<p style="text-align: center">  $$ e = \frac{W_{líquido}}{Q_{h}} = \frac{W_{turbina}- W_{bomba}}{Q_{h}}$$ </p> 

```python
#Eficiência Rankine
eta_th = (w_t-w_p)/q_H*100
print(f"Por fim, a eficiência térmica do sistema é de {round(float(eta_th),1)}%.")
```

<h3> Discussão Teórica: Análise da Eficiência do Ciclo de Carnot </h3>

<p style="text-align: justify"> O Ciclo de Carnot descreve a máxima eficiência térmica possível de uma máquina térmica sem quebrar as leis da natureza. O funcionamento deste ciclo consite em um sistema ou motor que transfere energia na forma de calor entre dois reservatórios térmicos a temperaturas baixas e altas, sendo que uma parte dessa energia é convertida em trabalho realizado pelo sistema. </p>

<p style="text-align: justify"> De modo geral, o ciclo atual em quatro processos como pode ser visto na Figura 4 que apresenta o ciclo termodinâmico de Carnot (similar ao Rankine), observa-se dois processos isotérmicos e dois processos isentrópicos: </p>

- Processo isotérmico: Os processos 1-2 e 3-4 são processos isotérmicos.
- Processo isentrópico: Os processos 2-3 e 4-1 são processos isentrópicos.
    
<p style="text-align: center"><img src="https://github.com/monocas/Redes-Neurais-e-Algoritmos-Geneticos/assets/106678040/8d267952-eaf8-464d-b55e-3c210acf4235" witdh = "300", height = "300" ></p>
<p style="text-align: center"> Figura 4. Ciclo termodinâmico do Ciclo de Carnot. <p>

<p style="text-align: justify"> No que diz respeito a sua eficiência, ela é determinada exclusivamente pelas temperaturas das duas fontes de calor. Ressalta-se que esta eficiência é uma característica intrínseca do ciclo de Carnot e não depende das propriedades específicas da substância de trabalho ou do mecanismo da máquina térmica. Ela é expressa matematicamente pela fórmula: </p>

$$ η = 1 - \frac{Tc}{Th}$$
    
 <p style="text-align: justify"> sendo: </p>
 
- Tc é a temperatura da fonte fria;
- Th é a temperatura da fonte quente. 

<p style="text-align: justify"> A eficiência do ciclo de Carnot é sempre menor que 1 e é a eficiência máxima possível para qualquer ciclo térmico operando entre essas duas temperaturas. Vamos analisar o ciclo de Carnot com os nossos valores de temperatura obtidos no experimento?</p>


```python
#Eficiência Carnot
#Constantes:
Tc = 318.96 # Kelvin
Th = 568.16 # Kelvin

#Fórmula
eta_th_carnot =  1 - (Tc/Th)

print(f"Por fim, a eficiência térmica do sistema é de {round(float(eta_th_carnot*100),1)}%.")
```

<h3> Comparação de Eficiências </h3>

<p style="text-align: justify"> A eficiência do ciclo de Carnot foi de 43.9%, enquanto a do Ciclo de Rankine foi de 35.6 %. Era esperado o Ciclo de Carnot apresentar um maior valor de eficiência devido suas propriedades teóricas termodinâmicas que promovem a reversibilidade do sistema - podem ser revertidos sem perda; por sua vez, o ciclo de Rankine é baseado em processos reais, que podem envolver irreversibilidades, como perdas de calor e atrito. Geralmente, Carnot é apenas usado como fim de comparação teórica para se ver a eficiência dos ciclos propostos. </p>

<p style="text-align: justify"> Outras diferenças básicas entre os sistemas é que no ciclo de Carnot o calor é adicionado e rejeitado a temperatura constante, enquanto no ciclo de Rankine, eles ocorrem com a pressão constante. A substância de trabalho também é diferente visto que Rankine utiliza de água, enquanto o Carnot utiliza o ar. </p>
