<h2 align="center"> <b> PLANO DE PROJETO: TERMOELÉTRICA MOANA S.A. </b>  </h3>

<p align="justify"> 
Termoelétricas produzem energia a partir do ciclo de potência a vapor, sendo o vapor gerado pela queima de combustíveis que fazem girar turbinas convertendo este trabalho em energia. O processo que faz essa produção de energia é denominado de Ciclo Rankine que aproveita o vapor que sai das turbinas condensando-o e transformando-o em líquido para realizar o ciclo repetidamente.</p>
<p align="justify"> 
Em síntese, ele possibilita por meio de ciclos termodinâmicos que o trabalho mecânico seja extraído de um fluido à medida que ele se move entre uma fonte de calor e um dissipador de calor. A fim de se ter essa energia é necessário ter uma boa queima de combustíveis com o intuito de promover uma grande quantidade de calor e energia, desse modo, o questionamento que se perpetua é:</p>
<p align="center"> 
"Como determinar bons combustíveis para a geração de energia?"

De modo a determinar bons combustíveis que gerem energia, um dos conceitos bases para esta definição é a entalpia e o diagrama de fases dos compositos.</p>


<h4 p align="justify"> <b> Entalpia </b> </p> </h4>
<p align="justify">  Entalpia pode ser compreendida como a quantidade de calor (energia) recebida ou absorvida no sistema por conta de uma mudança no próprio sistema ou pela quantidade de calor expelido pelo sistema como resultado da alteração deste. É um conceito muito utilizado em medições em sistemas biológicos, químicos e físicos a uma pressão constante fornecida pela atmosfera ambiente. </p>

<p align="justify">  De acordo com o  Schroeder et. al Addison Wesley Longman(1999), a ideia de processos de pressão constante é muito utilizada no dia a dia comercial e laboratorial em estudos de compressão-expansão e devido esta utilização constante, ao longo dos anos aprende-se algumas abordagens mais convenientes para a determinação destes, isto é,  podemos ao invés de pensar na energia do sistema, refletir a respeito do trabalho necessário para a geração do sistema que é expresso pelo produto da pressão pelo espaço (P x V) e quando adicionamos energia a este produto, obtemos a Entalpia, denotada por H: </p>

<p align="center"><img src="https://user-images.githubusercontent.com/106678040/226898642-1ad9644b-136b-4297-8ad1-3540bc9d480d.png"></p>
Fonte: David Schroeder (1999).
<p align="justify">
Essa mensuração é a H total do sistema que você precisaria obter para criar o sistema do nada e colocá-lo nesse ambiente. </p>


<p align="justify">A fim de entender a utilidade da Entalpia, pense em situações que ocorram uma mudança no sistema - reação de produtos químicos ou surgimento de calor -  enquanto a pressão é sempre mantida de modo constante. Por meio do cálculo de entalpia, conseguimos determinar a energia liberada em forma de calor por meio da movimentação das moléculas de uma reação quimíca transformada pelo produto desta.
  </p>
<p align="justify">
Existem alguns tipos de entalpia a serem utilizados, no presente momento, iremos dissertar a respeito de um tipo muito importante para a geração de calor: a entalpia de combustão.  </p>
<br>
<h4 p align="justify"> <b> Entalpia  de Combustão </b> </p> </h4>

<p align="justify">
Entalpia de combustão pode ser compreendida como uma energia liberada no modo de calor por meio de uma combustão de 1 mol de substância. Esta entalpia pode  se usada para comparar quais combustíveis ou substâncias liberam mais energia quando são queimados. </p>

<p align="justify">
Dada esta categorização e pensando no ciclo de Rankine, o método para a definição dos combustíveis para a mistura é baseada em uma proposta principal: compositos com alta entalpia de combustão. Desse modo, seleciona-se como mistura: Methanol e Ethyl Acetato. </p>
<br>
<h4 p align="justify"> <b> Mistura: Methanol + Ethyl Acetate </b> </p> </h4>
<p align="justify">
O Methanol(CH4O) é um alcoól volátil utilizado para inúmeros produtos do dia a dia, desde plástico até materiais de construção. Atualmente, ele também é utilizado como um recurso de energia limpa, sendo utilizado como um combustível para abastecer carros, navios, caldeiras e até fogões. </p>

<p align="justify"> Por sua vez, o Ethyl Acetate é um composto orgânico muito utilizado na indústria de solvente como solvente ou diluente e é formado pela composição entre o ácido acético e o etanol. Ademais, ele também é utilizado como combustíveis e na produção de plásticos. </p>

<p align="justify"> Dada a breve descrição dos compostos, vimos que esta mistura é composta por dois compostos utilizados como combustíveis o que contribui para a formação de combustão destes e consequentemente, a liberação de energia. Ademais, a entalpia destes apresenta um valor médio-alto de excecução, sendo estes Methanol (-715 kJ/mol) e Ethyl acetate(-478 kJ/mol) - os valores são negativos visto que representam a liberação da energia.</p>
